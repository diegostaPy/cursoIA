#ARIMA 
significa Media Mvil Integrada Auto-Regresiva. Hay tres nmeros enteros (p, d, q) que se utilizan para parametrizar modelos ARIMA. Por eso, un modelo ARIMA no estacional se denota con ARIMA(p, d, q):
<ul>
<li><strong>p</strong> es el nmero de trminos autorregresivos (parte AR). Permite incorporar el efecto de valores pasados a nuestro modelo. Intuitivamente, esto sera similar a afirmar que es probable que haga calor maana si ha estado caliente los ltimos 3 das.</li>
<li><strong>d</strong> es el nmero de diferencias no estacionales necesarias para la estacionariedad. Intuitivamente, esto sera similar a afirmar que es probable que maana haya la misma temperatura si la diferencia de temperatura en los ltimos tres das ha sido muy pequea.</li>
<li><strong>q</strong> es el nmero de errores de pronstico retrasados en la ecuacin de prediccin (parte MA). Esto nos permite establecer el error de nuestro modelo como una combinacin lineal de los valores de error observados en puntos de tiempo anteriores en el pasado.</li>
</ul>

Cuando se trata de efectos estacionales, como en nuestro ejemplo, se utiliza ARIMA estacional, que se denota como ARIMA(p,d,q)(P,D,Q)s. Aqu, (p, d, q) son los parmetros no estacionales descritos anteriormente, (<strong>P, D, Q</strong>) siguen la misma definicin pero se aplican al componente estacional de la serie temporal. El trmino <strong>s</strong> es la periodicidad de la serie temporal.

Aunque en este caso est claro que s=12, cmo fijamos los dems parmetros?

Se basa bastante en la experiencia. Existen numerosas prcticas recomendadas que pueden seguirse para identificar los modelos ARIMA, como por ejemplo
http://people.duke.edu/~rnau/arimrule.htm.

Aqu utilizamos la bsqueda de cuadrcula sobre todas las combinaciones posibles de valores de parmetros dentro de un rango predefinido de valores (muy inspirado en https://www.digitalocean.com/community/tutorials/a-guide-to-time-series- pronstico-con-arima-en-python-3).

$statsmodels.tsa.statespace.sarimax.SARIMAXResults$ devuelve valores de AIC (Criterio de Informacin de Akaike) y BIC (Criterio de Informacin de Bayes) que pueden minimizarse para seleccionar el modelo que mejor se ajuste. Utilizamos el valor AIC, que estima la informacin perdida cuando se utiliza un modelo determinado para representar el proceso que genera los datos. Al hacerlo, se ocupa del compromiso entre la bondad del ajuste del modelo y la complejidad del propio modelo.



```python
# Define the d and q parameters to take any value between 0 and 1
q = d = range(0, 2)
# Define the p parameters to take any value between 0 and 3
p = range(0, 4)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

We select a subset of the data series as training data, say the first 11 years. Our goal is to predict the last year of the series based on this input.

```python
train_data = data['1949-01-01':'1959-12-01']
test_data = data['1960-01-01':'1960-12-01']
```

```python
warnings.filterwarnings("ignore") # specify to ignore warning messages

AIC = []
SARIMAX_model = []
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(train_data,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('SARIMAX{}x{} - AIC:{}'.format(param, param_seasonal, results.aic), end='\r')
            AIC.append(results.aic)
            SARIMAX_model.append([param, param_seasonal])
        except:
            continue
```

```python
print('The smallest AIC is {} for model SARIMAX{}x{}'.format(min(AIC), SARIMAX_model[AIC.index(min(AIC))][0],SARIMAX_model[AIC.index(min(AIC))][1]))
```

```python
# Let's fit this model
mod = sm.tsa.statespace.SARIMAX(train_data,
                                order=SARIMAX_model[AIC.index(min(AIC))][0],
                                seasonal_order=SARIMAX_model[AIC.index(min(AIC))][1],
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()
```

Una vez que el modelo ha sido ajustado, podemos verificar si hace lo que esperbamos y si se violan las suposiciones que hicimos. Para hacer esto, podemos usar el mtodo $plot\_diagnostics$.

```python
results.plot_diagnostics(figsize=(20, 14))
plt.show()
```

En los grficos anteriores, podemos observar que los residuos no estn correlacionados (grfico inferior derecho) y no muestran ninguna estacionalidad obvia (grfico superior izquierdo). Adems, los residuos y aproximadamente normalmente distribuidos con media cero (grfico superior derecho). El diagrama qq en la parte inferior izquierda muestra que la distribucin ordenada de residuos (puntos azules) sigue aproximadamente la tendencia lineal de las muestras tomadas de una distribucin normal estndar con N (0, 1). Nuevamente, esta es una fuerte indicacin de que los residuos se distribuyen normalmente.

Resultados

1) En prediccin de muestra con pronstico de 1 paso adelante del ltimo ao (1959). En este caso, el modelo se usa para predecir los datos sobre los que se construy el modelo. El pronstico de 1 paso adelante implica que cada punto pronosticado se usa para predecir el siguiente.

```python
pred0 = results.get_prediction(start='1958-01-01', dynamic=False)
pred0_ci = pred0.conf_int()
```

2) En prediccin muestral con prediccin dinmica del ltimo ao (1959). Una vez ms, el modelo se utiliza para predecir los datos sobre los que se construy el modelo.

```python
pred1 = results.get_prediction(start='1958-01-01', dynamic=True)
pred1_ci = pred1.conf_int()
```

Cuando establece dynamic=True, el modelo predice continuamente un paso adelante (t+1) y luego, para la prediccin del segundo paso adelante (t+2), agrega el valor predicho (t+1) a los datos, vuelve a ajustar el modelo a partir de los nuevos datos ampliados, se realiza un pronstico de 2. paso adelante. Esto se llama prediccin fuera de la muestra.

Cuando establece dinmico = Falso, el modelo predice secuencialmente un paso adelante usando el valor verdadero del paso de tiempo anterior en lugar de usar el valor predicho. Esto se llama prediccin en la muestra.

3) Pronstico "verdadero" de datos fuera de muestra. En este caso, se le pide al modelo que prediga datos que no ha visto antes.

```python
pred2 = results.get_forecast('1962-12-01')
pred2_ci = pred2.conf_int()
print(pred2.predicted_mean['1960-01-01':'1960-12-01'])
```

```python
ax = data.plot(figsize=(20, 16))
pred0.predicted_mean.plot(ax=ax, label='1-step-ahead Forecast (get_predictions, dynamic=False)')
pred1.predicted_mean.plot(ax=ax, label='Dynamic Forecast (get_predictions, dynamic=True)')
pred2.predicted_mean.plot(ax=ax, label='Dynamic Forecast (get_forecast)')
ax.fill_between(pred2_ci.index, pred2_ci.iloc[:, 0], pred2_ci.iloc[:, 1], color='k', alpha=.1)
plt.ylabel('Monthly airline passengers (x1000)')
plt.xlabel('Date')
plt.legend()
plt.show()
```

Mirando la figura, el modelo parece hacer un buen trabajo al modelar la serie de tiempo. Las lneas azul y prpura estn, como era de esperar, muy cerca de la verdad del fondo rojo. Lo que es ms interesante es la lnea gris, la prediccin fuera de muestra. Para una serie de tiempo tan simple, el modelo ARIMA puede pronosticar los valores de 1960 con precisin.

Para cuantificar la precisin de la prediccin para 1960, podemos calcular mtricas como el error absoluto medio, el error cuadrtico medio o el error cuadrtico medio. Todas estas son mtricas absolutas, por lo tanto, dependen de la escala. Si bien son tiles y se usan ampliamente para comparar diferentes mtodos en un mismo conjunto de datos, aqu sera ms til expresar el error en relacin con la magnitud de la serie temporal que estamos tratando de predecir. Una medida de error porcentual comnmente utilizada es el error porcentual absoluto medio (MAPE). Tiene algunos inconvenientes en casos especiales (en particular, puede llevar a la divisin por cero) y se han propuesto medidas mejoradas, vase, p. https://www.otexts.org/fpp/2/5 para obtener una buena descripcin general. Sin embargo, para este ejemplo nos quedaremos con MAPE.

```python
prediction = pred2.predicted_mean['1960-01-01':'1960-12-01'].values
# flatten nested list
truth = list(itertools.chain.from_iterable(test_data.values))
# Mean Absolute Percentage Error
MAPE = np.mean(np.abs((truth - prediction) / truth)) * 100

print('The Mean Absolute Percentage Error for the forecast of year 1960 is {:.2f}%'.format(MAPE))
```
