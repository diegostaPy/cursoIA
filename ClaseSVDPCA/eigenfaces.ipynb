```python
import matplotlib.pyplot as plt #plot import
import matplotlib.colors  #color import
import numpy as np  #importing numpy
from PIL import Image #importing PIL to read all kind of images
#from PIL import ImageTk
import glob
```

```python

def mostrarImagenes(img):
    size=100, 100
    #fig1=plt.figure('name')
    fig1, axes_array = plt.subplots(5, 5)
    fig1.set_size_inches(5,5)
    count=0
    for x in range(5):
        for y in range (5):
            image_plot=axes_array[x][y].imshow(img[count],cmap=plt.cm.gray)
            axes_array[x][y].axis('off')
            count=count+1
    plt.show()
```

```python
lista_imagenes = []
for archivo in glob.glob('Train/*.jpg'): # leer archivos
    imagen=Image.open(archivo)
    imagen_arr = np.asarray(imagen,dtype=float)/255.0
    lista_imagenes.append(imagen_arr)
```

```python
mostrarImagenes( lista_imagenes)
```

```python
np.size(lista_imagenes[0])
```

```python
array_imagenes=np.asarray(lista_imagenes)
```

```python
media=np.mean(array_imagenes,axis=0)
```

```python
plt.imshow(media)
plt.show()
# rostro medio
```

```python
B = np.subtract(array_imagenes, media)
```

```python
B=B.reshape(25,180625)
```

```python

U, s, V = np.linalg.svd(B, full_matrices=False)
Eigen_faces=[]
# recordando que V esta asociado a la espacio de columnas, las imagenes
# esta asociado a los pixeles, espacio de las filas
# recurde que 
for x in range(V.shape[0]):
    fig=np.reshape(V[x],(425,425))
    Eigen_faces.append(fig)
```

```python
mostrarImagenes( Eigen_faces)
```
